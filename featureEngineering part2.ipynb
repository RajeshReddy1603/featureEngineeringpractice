```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
a={
    "names":["amazon","asda","lidil","tesco","coop"],
     "values":["rajesh","maddi","today","tomorrow","ok"]
}
df=pd.DataFrame(a)
obj=LabelEncoder()
df["names_n"]=obj.fit_transform(df["names"])
df["values_n"]=obj.fit_transform(df["values"])
df
```
